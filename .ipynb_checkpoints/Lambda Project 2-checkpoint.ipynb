```python
import requests
from bs4 import BeautifulSoup
```

```python
data = requests.get('https://www.hrw.org/world-report/2019/country-chapters/united-states#80232e')
net = BeautifulSoup(data.text, 'html.parser')

leader_names = net.find(class_='BodyText')
leader_list = leader_names.findAll('a')

for leader in leader_list:
    print(leader_names.prettify())
```
